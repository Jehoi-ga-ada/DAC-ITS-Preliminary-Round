## We see that right now the dataset doesn't have a lot of missing values, so we will try to find some anomaly on the data

```python
set(raw_data_train['type_of_attack'].values)
```

```python
for col in raw_data_train.columns:
  print(col)
  print(raw_data_train[col].nunique())
  print(set(raw_data_train[col].values), "\n\n")
```

## Now we see that there is some anomaly in the data where the data contains * or 99999

We can print the column with anomaly

Columns with anomaly

```python
columns_with_anomalies = []
```

```python
for col in raw_data_train.columns:
    unique_values = set(raw_data_train[col].values)
    if "*" in unique_values or 99999 in unique_values:
        columns_with_anomalies.append(col)
print(columns_with_anomalies)
```

Columns with NaN

```python
columns_with_na = []
```

```python
for col in raw_data_train.columns:
    if raw_data_train[col].isna().sum() > 0:
        columns_with_na.append(col)
print(columns_with_na)
```

We should check that if there is any intersection of the columns with values of nan and columns with anomaly so that we can preprocess them further

```python
for col in columns_with_na:
    if col in columns_with_anomalies:
        print(col)
```

We can see that there is intersection between the missing values and anomaly columns

For now we will check for the missing values

```python
data_with_na = raw_data_train[columns_with_na]
data_with_na
```

```python
total_missing_values = row - data_with_na[data_with_na.isna().sum(1)==0].shape[0]
total_missing_values
```

now we check the percentage of the missing values

```python
print(round(total_missing_values/row*100, 2), "%")
```

we can see that the missing values is under 10%. This means we can drop the missing values since dropping it won't affect the data quality

```python
data_no_na = raw_data_train.dropna()
```

```python
row = data_no_na.shape[0]
row
```

```python
data_no_na
```

now we deal with the anomalies

```python
data_with_anomalies = data_no_na[columns_with_anomalies]
```

```python
data_with_anomalies.isna().sum()
```

We count the data with anomalies to decide wether will we drop it or impute it

```python
anomalies_occurrences_per_row = ((raw_data_train == '*') | (raw_data_train == '99999') | (raw_data_train == 99999)).sum(axis=1)>0
total_anomalies = anomalies_occurrences_per_row.replace({True:1, False:0}).sum()
anomalies_percentage = round(total_anomalies/row*100, 2)
print(f'Anomalies Percentage on the given dataset after dropping na: {anomalies_percentage}%')
```

As we can see that in the data we have 88% anomalies from the total data. This means we cannot simply drop the data we will need to impute the data

```python
binary_features = ['land', 'logged_in', 'root_shell', 'su_attempted', 'is_host_login', 'is_guest_login']
```

```python
for col in binary_features:
    print(data_with_anomalies[col].value_counts())
```

```python
data_no_na = data_no_na.drop(data_no_na[data_no_na.su_attempted == '2'].index, axis=0)
data_no_na
```

```python
for col in binary_features:
    print(data_no_na[col].value_counts())
```

we found out that the binary features has some out of the range values so we drop that values to ensure it still binary

```python
data_no_na[binary_features] = data_no_na[binary_features].replace({'*':'0', '99999': '1'})
for col in binary_features:
    print(data_no_na[col].value_counts())
```

now that we impute the anomalies within the binary featues, we can move on to the next variable typing 

```python
categorical = ['flag', 'service', 'protocol_type', 'type_of_attack']
```

```python
continous = []
for i in data_no_na.columns:
    if i not in binary_features and i not in categorical:
        continous.append(i)
```

```python
for col in continous:
    print(data_no_na[col].value_counts())
```

```python
percetage_feature = ['serror_rate',
'srv_serror_rate',
'rerror_rate',
'srv_rerror_rate',
'same_srv_rate',
'dst_host_same_srv_rate',
'dst_host_diff_srv_rate',
'dst_host_same_src_port_rate',
'dst_host_srv_diff_host_rate',
'dst_host_serror_rate' ,
'dst_host_srv_serror_rate',
'dst_host_rerror_rate',
'dst_host_srv_rerror_rate']
```

```python
data_no_na[percetage_feature] = data_no_na[percetage_feature].replace({'*':'0', '99999': '1'})
for col in percetage_feature:
    print(data_no_na[col].value_counts())
```
