## We see that right now the dataset doesn't have a lot of missing values, so we will try to find some anomaly on the data

```python
set(raw_data_train['type_of_attack'].values)
```

```python
for col in raw_data_train.columns:
  print(col)
  print(raw_data_train[col].nunique())
  print(set(raw_data_train[col].values), "\n\n")
```

## Now we see that there is some anomaly in the data where the data contains * or 99999

We can print the column with anomaly

Columns with anomaly

```python
columns_with_anomalies = []
```

```python
for col in raw_data_train.columns:
    unique_values = set(raw_data_train[col].values)
    if "*" in unique_values or 99999 in unique_values:
        columns_with_anomalies.append(col)
print(columns_with_anomalies)
```

Columns with NaN

```python
columns_with_na = []
```

```python
for col in raw_data_train.columns:
    if raw_data_train[col].isna().sum() > 0:
        columns_with_na.append(col)
print(columns_with_na)
```

We should check that if there is any intersection of the columns with values of nan and columns with anomaly so that we can preprocess them further

```python
for col in columns_with_na:
    if col in columns_with_anomalies:
        print(col)
```

We can see that there is intersection between the missing values and anomaly columns

For now we will check for the missing values

```python
data_with_na = raw_data_train[columns_with_na]
data_with_na
```

```python
total_missing_values = row - data_with_na[data_with_na.isna().sum(1)==0].shape[0]
total_missing_values
```

now we check the percentage of the missing values

```python
print(round(total_missing_values/row*100, 2), "%")
```

we can see that the missing values is under 10%. This means we can drop the missing values since dropping it won't affect the data quality

```python
data_no_na = raw_data_train.dropna()
```

```python
data_no_na
```
