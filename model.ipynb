```python
raw_data = pd.read_csv('clean_train_data.csv')
raw_data
```

```python
features_df = raw_data.iloc[:, :-1]
target_df = pd.DataFrame(raw_data['type_of_attack'])
```

Now we have to encode the target so we can use it for the machine learning algorithm

```python
from sklearn.preprocessing import LabelEncoder
```

```python
Encoder = LabelEncoder()
label_encoded_target = pd.DataFrame(np.array(Encoder.fit_transform(target_df)).T, columns=['type_of_attack'])
label_encoded_target
```

Now we have to encode the labels, for this dataset we will use mean encoding technique because the label has a lot of categorical value instead of binary, this method work well with supervised learning and classification problem.
