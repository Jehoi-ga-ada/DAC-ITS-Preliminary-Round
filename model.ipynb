```python
import pandas as pd
import numpy as np
```

```python
raw_data = pd.read_csv('clean_train_data.csv')
raw_data
```

```python
features_df = raw_data.iloc[:, :-1]
target_df = pd.DataFrame(raw_data['type_of_attack'])
```

Now we have to encode the target so we can use it for the machine learning algorithm

```python
from sklearn.preprocessing import LabelEncoder
```

Now we have to encode the target, for this dataset we will use mean encoding technique because the label has a lot of categorical value instead of binary, this method work well with supervised learning and classification problem.

```python
Encoder = LabelEncoder()
label_encoded_target = pd.DataFrame(np.array(Encoder.fit_transform(target_df)).T, columns=['type_of_attack'])
label_encoded_target
```

now we have to encode the categorical features

```python
categorical_columns = ['flag', 'service', 'protocol_type']
categorical_feature = features_df[categorical_columns]
categorical_feature
```

```python
pre_mean_encoded_df = pd.concat([categorical_feature, label_encoded_target], axis = 1)
pre_mean_encoded_df
```

```python
from category_encoders import TargetEncoder
```

```python
encoder = TargetEncoder()
mean_encoded_features = encoder.fit_transform(categorical_feature, label_encoded_target)
```

```python
features_df.columns.values
```

```python
for i in categorical_columns:    
    features_df = features_df.drop([i], axis = 1)
```

```python
features_df
```

```python
features_encoded = pd.concat([features_df, mean_encoded_features], axis = 1)
```

```python
features_encoded = features_encoded[['duration', 'protocol_type', 'service', 'flag', 'src_bytes',
       'dst_bytes', 'land', 'wrong_fragment', 'urgent', 'hot',
       'num_failed_logins', 'logged_in', 'num_compromised', 'root_shell',
       'su_attempted', 'num_root', 'num_file_creations', 'num_shells',
       'num_access_files', 'num_outbound_cmds', 'is_host_login',
       'is_guest_login', 'count', 'srv_count', 'serror_rate',
       'srv_serror_rate', 'rerror_rate', 'srv_rerror_rate',
       'same_srv_rate', 'diff_srv_rate', 'srv_diff_host_rate',
       'dst_host_count', 'dst_host_srv_count', 'dst_host_same_srv_rate',
       'dst_host_diff_srv_rate', 'dst_host_same_src_port_rate',
       'dst_host_srv_diff_host_rate', 'dst_host_serror_rate',
       'dst_host_srv_serror_rate', 'dst_host_rerror_rate',
       'dst_host_srv_rerror_rate']]
features_encoded
```

```python
label_encoded_target
```

Now that we encoded all of our categorical features and label, we will balance the data so the model does not learn to predict one category more than the others

```python
from imblearn.over_sampling import RandomOverSampler, SMOTE
import xgboost as xgb
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import classification_report
from imblearn.under_sampling import RandomUnderSampler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
```

```python
label_encoded_target.value_counts()
```

As you can see above we have a very imbalanced dataset, therefore we will try using 3 methods which is undersampling, oversampling, and smote. After that, we will compare the accuracy of each model with the given data

```python
preprocessed_data = pd.concat([features_encoded, label_encoded_target], axis = 1)
preprocessed_data
```

```python
X = preprocessed_data.drop(columns=['type_of_attack'])
Y = preprocessed_data['type_of_attack']
```

```python
Y.nunique()
```

We have 8 neigbors omggg

```python
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
smote = SMOTE(random_state=42)
x_train_smote_resampled, y_train_smote_resampled = smote.fit_resample(x_train, y_train)
```

```python
undersampler = RandomUnderSampler(random_state=42)
x_train_undersampled_resampled, y_train_undersampled_resampled = undersampler.fit_resample(x_train, y_train)
```

```python
oversampler = RandomOverSampler(random_state=42)
x_train_oversampled_resampled, y_train_oversampled_resampled = oversampler.fit_resample(x_train, y_train)
```

```python
knn_smote = KNeighborsClassifier(n_neighbors=8)
knn_smote.fit(x_train_smote_resampled, y_train_oversampled_resampled)
```

```python
knn_undersampled = KNeighborsClassifier(n_neighbors=8)
knn_undersampled.fit(x_train_undersampled_resampled, y_train_undersampled_resampled)
```

```python
knn_oversampled = KNeighborsClassifier(n_neighbors=8)
knn_oversampled.fit(x_train_oversampled_resampled, y_train_oversampled_resampled)
```

```python
smote_y_predict = knn_smote.predict(x_test)
undersampled_y_predict = knn_undersampled.predict(x_test)
oversampled_y_predict = knn_oversampled.predict(x_test)
```

```python
smote_report = accuracy_score(y_test, smote_y_predict)
undersampled_report = accuracy_score(y_test, undersampled_y_predict)
oversampled_report = accuracy_score(y_test, oversampled_y_predict)
```

```python
print("KNN using smote data :", smote_report)
print("KNN using undersampled data :", undersampled_report)
print("KNN using oversampled data :", oversampled_report)
```

we can see the oversampling the data would be appropriate since we have a small data set, and the difference between the classes are very high

```python
models = {
    'XGBoost (Logistic)': xgb.XGBClassifier(objective='multi:softmax', num_class=8),
    'Decision Tree': DecisionTreeClassifier(),
    'Random Forest': RandomForestClassifier(),
    'Naive Bayes (Gaussian)': GaussianNB()
}
```

```python
for model_name, model in models.items():
    # Train the model
    model.fit(x_train_undersampled_resampled, y_train_undersampled_resampled)

    # Predict on the test set
    y_pred = model.predict(x_test)

    # Calculate accuracy
    accuracy = accuracy_score(y_test, y_pred)
    print(f"{model_name} Accuracy: {accuracy}")
```

```python
for model_name, model in models.items():
    # Train the model
    model.fit(x_train_oversampled_resampled, y_train_oversampled_resampled)

    # Predict on the test set
    y_pred = model.predict(x_test)

    # Calculate accuracy
    accuracy = accuracy_score(y_test, y_pred)
    print(f"{model_name} Accuracy: {accuracy}")
```

```python
for model_name, model in models.items():
    # Train the model
    model.fit(x_train_smote_resampled, y_train_smote_resampled)

    # Predict on the test set
    y_pred = model.predict(x_test)

    # Calculate accuracy
    accuracy = accuracy_score(y_test, y_pred)
    print(f"{model_name} Accuracy: {accuracy}")
```

we use XGB as the main model cause its works well and give better accuracy 

```python
XGB = xgb.XGBClassifier(objective='multi:softmax', num_class=8)
XGB.fit(x_train_oversampled_resampled, y_train_oversampled_resampled)
y_pred = XGB.predict(x_test)
accuracy = accuracy_score(y_test, y_pred)
print(f"XGB Accuracy: {accuracy}")
```
